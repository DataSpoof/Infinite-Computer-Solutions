```python
# ---------------- NEWS ----------------
async def fetch_news(session):
    url = f"https://newsapi.org/v2/top-headlines?country=us&apiKey={NEWS_API_KEY}"
    async with session.get(url, headers=HEADERS) as response:
        data = await response.json()
        articles = data.get("articles", [])[:5]
        print("\n📰 Top News Headlines:")
        for article in articles:
            print("-", article["title"])
        return articles

```

```python
# ---------------- STOCK ----------------
async def fetch_stock(session, symbol="AAPL"):
    try:
        url = f"https://query1.finance.yahoo.com/v8/finance/chart/{symbol}"
        async with session.get(url, headers=HEADERS) as response:
            if response.status != 200:
                print(f"Stock API Error: {response.status}")
                return None

            data = await response.json()
            # data= await asyncio.wait_for(response.json(),timeout=5)
            result = data["chart"]["result"][0]
            price = result["meta"]["regularMarketPrice"]
            print(f"\n📈 {symbol} Current Price:", price)
            return price

    except Exception as e:
        print("Stock fetch failed:", e)
        return None

```

```python
# ---------------- MAIN ----------------
async def main():
    start = time.time()

    async with aiohttp.ClientSession() as session:
      # task is created in the coroutines
        news_task = asyncio.create_task(fetch_news(session))
        stock_task = asyncio.create_task(fetch_stock(session))
      # gathers all the information
        news, stock = await asyncio.gather(news_task, stock_task)

    end = time.time()
    print(f"\n⏱ Total Time Taken: {end-start:.2f} seconds")

```

```python
#asyncio.run(main())  # if you have .py file then use the first line
await main()
```
