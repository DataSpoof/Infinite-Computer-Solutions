# Inheritance

* you inherit the properties (variables and methods) from your parents

```python
class building:
  def __init__(self,name,price,year):
    self.name=name
    self.price=price
    self.year=year

  def print_details(self):
    print(f"the name is {self.name} and price is {self.price} and year{self.year}")
```

```python
# child class
class room(building):
  def __init__(self,room_no,name, price, year):
    self.room_no=room_no
    super().__init__(name,price,year) # Super keyword is used for inheritance

  def updated_details(self):
    print(f"the name {self.name} and price {self.price} and year{self.year} and room no is {room_no}")
```

# Method resolution order (MRO) ****

* it is used in multiple inheritance

* grand parent, parent, you (granparent, parent)

* resolves the ambiguity in multiple inheritance

* ensure each class is visited only once

* it follows the sequence to search for methods and attribes inheritance correctly

```python
class A: # grandparent
  def print(self):
    print("grandparent")

class B(A): # parent
  pass

class C(A): # bua (father-sister)
  pass

class D(B,C): #you(father, and his sister)
  pass

d= D()
d.print()
```

D--- B---C---A--object

```python
print(D.mro())
```

# Abstraction

* showing you the big picture without worrying about the internal details

* it hides the internal methods only show you the big picture

```python
from abc import ABC,abstractmethod

class Polygon(ABC):

  @abstractmethod # it hides the internal implementation from child
  def area(self):
    pass

  @abstractmethod
  def perimeter(self):
    pass
```

```python
class Triangle(Polygon):

  def area(self):
    print("area of triangle")

  def perimeter(self):
    return 5
```

```python
R= Triangle()
R.area()
```

# Encapsulation

* It introduces private variables
* If it is private it should not be accessed directly from outside the class

* Getter is used to set the values of private variables
* setter is used to set the values of private variables
* it helps us in providing security by restricting the access to other person

```python
class BankAccount():
  def __init__(self,name,balance):
    self.name=name
    self.__balance=balance

  def get_balance(self):
    return self.__balance # in order to access the private variables

  def deposit(self,amount):
    self.__balance+=amount
```

```python
result= BankAccount("Abhi",5000)
```

```python
result.get_balance()
```

```python
result.deposit(1000)
```

```python
result.get_balance()
```
