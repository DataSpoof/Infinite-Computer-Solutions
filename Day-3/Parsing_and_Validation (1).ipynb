# Step 2: Import Pydantic and Define a Model


```python
from pydantic import BaseModel

class Person(BaseModel):
    name: str
    age: int
    email: str
```

# Bonus: Automatic Type Conversion


```python
p3 = Person(name="Alice", age="25", email="alice@example.com")
print(p3)
```

```python
from pydantic import BaseModel,field_validator

Allowed={'Monday','Tuesday'}

# dont give llm all the permission in sql (necessary permission will be given)

class User(BaseModel):
  dates:str

  @field_validator('dates')
  def check_dates(cls,v):
    if v not in Allowed:
      raise ValueError(f"Invalid date {v}")
    return v
```

```python
User(dates="Monday")
```

```python
User(dates="Sunday")
```

# Dataclass

Dataclass automatically creates:

* __init__

* __repr__ (representation of an object)

* __eq__ (equality operator)

* post_init (to post checking also)

You write less code ✨

## Without dataclass

```python
class User:
    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age

    def __repr__(self): # representation of the object
        return f"User(name='{self.name}', age={self.age})"

    def __eq__(self, other): # equality
        return self.name == other.name and self.age == other.age
```

```python
from dataclasses import dataclass

@dataclass # a decorator which will add the functionality to my existing code
class User:
    name: str
    age: int
    id: int
```

```python
# Creating objects
user1 = User("Abhishek", 30,345678)
user2 = User("Rahul", 30,43535)
user3 = User("Abhishek", 25,4343545)
```

```python
# Accessing attributes
print(user1.name)   # Abhishek
print(user1.age)    # 25
print(user1.id)
```

```python
# __repr__ auto-generated
print(user1)

```

```python
# __eq__ auto-generated (value comparison)
print(user1 == user3)  # True
print(user1 == user2)  # False
```

# Private attribute in dataclass

```python
from dataclasses import dataclass
@dataclass
class User:
    name: str
    __age: int   # private
```

```python
u = User("Abhishek", 25)
print(u.__age)
```

```python
print(u._User__age)
```
