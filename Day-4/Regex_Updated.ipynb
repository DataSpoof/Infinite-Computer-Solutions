## Regex Flags

```python
import re

text="Hello\nhello\nHELLO"

print(re.findall("hello", text))
print(re.findall("hello", text, re.IGNORECASE))
print(re.findall("^hello", text, re.MULTILINE|re.IGNORECASE))
```

## match(), search(), finditer()

```python
import re

text="My roll number is 101"

print(re.match(r"\d+", text))
print(re.search(r"\d+", text))

for m in re.finditer(r"\d+", text):
    print(m.group(), m.span())
```

## findall() and split()

```python
import re

text="Python,Java,C++,SQL"

print(re.findall(r"[A-Za-z+]+", text))
print(re.split(r",", text))
```

## Match Object: group(), start(), end(), span()

```python
import re

text="Order ID: 12345"

m=re.search(r"\d+", text)

if m:
    print("group:",m.group())
    print("start:",m.start())
    print("end:",m.end())
    print("span:",m.span())
```
